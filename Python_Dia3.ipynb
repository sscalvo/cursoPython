def nombre_funcion():  # Usar nombres descriptivos 💥

```
Una vez creada la función podremos llamarla (o invocarla) de igual manera que hemos hecho hasta ahora: Simplemente escribimos el nombre de la función seguida de 2 paréntesis:  ```nombre_funcion()```






## Funciones que aceptan parámetros
Cuando invocamos la función saludar, se imprime "Hola Mundo!". La función no necesita de ninguna información adicional para cumplir su cometido:

```
def saludar():  
    print("Hola Mundo!!")
```
Sin embargo, podemos querer que la función salude a una persona específica. Para ello, es muy útil poder parametrizar la función:



```
# La función acepta ahora un parámetro llamado nombre del cual hará uso en el código
def saludar_persona(nombre): 
    print(f"Hola {nombre}!!")
```
La manera de invocar a ambas funciones es:

```
saludar()               # imprime "Hola Mundo!!"
saludar_persona("Luis") # imprime "Hola Luis"
```






## Cuando las funciones devuelven un valor..
Ambas funciones (saludar y saludar_persona) no devuelven nada. En caso de intentar asignarlas a una variable para recoger el valor de retorno, veremos que es  **None**


```
valor = saludar() # imprime "Hola Mundo!!"
print(valor) # valor es None
```

Si la función no tiene la instrucción **return** en su código, el valor de la función es **None**. Sin embargo, podemos hacer que las funciones devuelvan un valor haciendo uso de **return**:


```
def factura(cantidad, iva):              # Acepta dos parámetros
   total = cantidad + cantidad*iva/100   # Hacemos los cálculos necesarios
   return total    
```
El código invocante es responsable de recoger el valor de retorno, para su posterior uso: 


```
valor = factura(100, 21) # invocamos a la función y la asignamos a una variable
print(f"El valor de la factura de 100€ y 21% de IVA es {valor}") 

```






## Funciones que aceptan parámetros
Una función puede aceptar uno o más parámetros:


```
def factura(cantidad, iva): # Dos parámetros
    return cantidad + cantidad*iva/100 
```
Si invocamos la función pasando no solo el valor del parámetro sino también el nombre (argumentos  clave=valor), entonces el orden de los parámetros no importa:


```
# Si pasamos sólo el valor hay que respetar el orden
impuestos = factura(840, 21) 

# Si indicamos el nombre del parámetro ambas formas son equivalentes
impuestos = factura(iva=21, cantidad=840) 
impuestos = factura(cantidad=840, iva=21) 

```




## Funciones cuyos parámetros tienen un valor por defecto
Cuando creamos (definimos) una función, podemos establecer valores por defecto a los parámetros, de tal manera que, si a la hora de invocar dicha función omitimos alguno de los parámetros, el código usará ese valor por defecto.


```
def factura(cantidad, iva=21): 
    return cantidad + cantidad*iva/100 

# ¿Qué sucede en cada una de estas invocaciones?
print( factura(100)     )
print( factura(100, 21) )
print( factura(21, 100) )
print( factura(21)      )
print( factura(iva=21)  )
print( factura(iva=21, cantidad=100) )
print( factura(cantidad=100, iva=21) )

```



## Funciones con un número variable de parámetros
Puede suceder que nuestras funciones sean tan genéricas que no sepamos de antemano cuántos parámetros debe esperar recibir. No problem! Podemos añadir un asterisco (*) delante del nombre del parámetro para indicar que aceptamos un número de parámetros desconocido y variable.


```
def sumatoria(*numeros): 
    total = 0
    for numero in numeros:
        total = total + numero
    return total

# 
print(sumatoria(1, 2, 3))    # 6
print(sumatoria(1, 2, 3, 4)) # 10
```
En caso de necesitar usar parámetros fijos, podemos combinar ambos:



```
def sumatoria_de_potencias(potencia, *numeros): 
    total = 0
    for numero in numeros:
        total = total + numero**potencia
    return total

print(sumatoria_de_potencias(3, 1, 2, 3))    # 36
print(sumatoria_de_potencias(2, 1, 2, 3, 4)) # 30
```






## Funciones con un número variable de parámetros (con nombre)
En el ejemplo anterior hemos invocado a la función factura con named arguments (pasábamos no solo el valor sino tambien el nombre del argumento):


```

# Pasando una lista (mutable)
def modifica(dato):
    dato = [7] # intentamos modificar la referencia. En realidad estamos creando una
    # nueva lista y asignándolo a 'dato', pero no modificamos la referencia original 
    # dato.append(7) # Esto si cambiaría los datos, el invocante también lo vería 

lista = [3]
modifica(lista)
print(lista) # [3], no ha cambiado, ya que la referencia 'dato' fue pasada por valor
```
.. y de tipo inmutable.


```
# Pasando una tupla (inmutable)
def modifica(dato):
    dato = (7,)   # intentamos modificar la referencia 
    # dato.append(7) # no hay metodos de modificación en los inmutables!! 

tupla= (3,)
modifica(tupla)
print(tupla) # (3,) no ha cambiado, ya que la referencia 'dato' fue pasada por valor
```
