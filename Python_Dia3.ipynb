def saludar():  
    print("Hola Mundo!!")
```
Sin embargo, podemos querer que la funci√≥n salude a una persona espec√≠fica. Para ello, es muy √∫til poder parametrizar la funci√≥n:



```
# La funci√≥n acepta ahora un par√°metro llamado nombre del cual har√° uso en el c√≥digo
def saludar_persona(nombre): 
    print(f"Hola {nombre}!!")
```
La manera de invocar a ambas funciones es:

```
saludar()               # imprime "Hola Mundo!!"
saludar_persona("Luis") # imprime "Hola Luis"
```






## Cuando las funciones devuelven un valor..
Ambas funciones (saludar y saludar_persona) no devuelven nada. En caso de intentar asignarlas a una variable para recoger el valor de retorno, veremos que es  **None**


```
valor = saludar() # imprime "Hola Mundo!!"

```

Si la funci√≥n no tiene la instrucci√≥n **return** en su c√≥digo, el valor de la funci√≥n es **None**. Sin embargo, podemos hacer que las funciones devuelvan un valor haciendo uso de **return**:


```
def factura(cantidad, iva):              # Acepta dos par√°metros
   total = cantidad + cantidad*iva/100   # Hacemos los c√°lculos necesarios
   return total    
```
El c√≥digo invocante es responsable de recoger el valor de retorno, para su posterior uso: 


```
valor = factura(100, 21) # invocamos a la funci√≥n y la asignamos a una variable
print(f"El valor de la factura de 100‚Ç¨ y 21% de IVA es {valor}") 

```






## Funciones que aceptan par√°metros
Una funci√≥n puede aceptar uno o m√°s par√°metros:


```

```



## Funciones con un n√∫mero variable de par√°metros
Puede suceder que nuestras funciones sean tan gen√©ricas que no sepamos de antemano cu√°ntos par√°metros debe esperar recibir. No problem! Podemos a√±adir un asterisco (*) delante del nombre del par√°metro para indicar que aceptamos un n√∫mero de par√°metros desconocido y variable.


```
def sumatoria(*numeros): 
    total = 0
    for numero in numeros:
        total = total + numero
    return total

# 
print(sumatoria(1, 2, 3))    # 6
print(sumatoria(1, 2, 3, 4)) # 10
```
En caso de necesitar usar par√°metros fijos, podemos combinar ambos:



```
def sumatoria_de_potencias(potencia, *numeros): 
    total = 0
    for numero in numeros:
        total = total + numero**potencia
    return total

print(sumatoria_de_potencias(3, 1, 2, 3))    # 36
print(sumatoria_de_potencias(2, 1, 2, 3, 4)) # 30
```






## Funciones con un n√∫mero variable de par√°metros (con nombre)
En el ejemplo anterior hemos invocado a la funci√≥n factura con named arguments (pas√°bamos no solo el valor sino tambien el nombre del argumento):


```

# Pasando una lista (mutable)
def modifica(dato):
    dato = [7] # intentamos modificar la referencia. En realidad estamos creando una
    # nueva lista y asign√°ndolo a 'dato', pero no modificamos la referencia original 
    # dato.append(7) # Esto si cambiar√≠a los datos, el invocante tambi√©n lo ver√≠a 

lista = [3]
modifica(lista)
print(lista) # [3], no ha cambiado, ya que la referencia 'dato' fue pasada por valor
```
.. y de tipo inmutable.


```
# Pasando una tupla (inmutable)
def modifica(dato):
    dato = (7,)   # intentamos modificar la referencia 
    # dato.append(7) # no hay metodos de modificaci√≥n en los inmutables!! 

tupla= (3,)
modifica(tupla)
print(tupla) # (3,) no ha cambiado, ya que la referencia 'dato' fue pasada por valor
```
